# `torch-dcr` Example Notebook

In this Notebook we will show how to use the `torch-dcr` library to compute the DCR (Distance to the Closest Representative) for a given dataset.

First, let's load the Adult dataset:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/rolester/AzureMLModelExplain/master/Adult%20Census%20Income%20Binary%20Classification%20dataset.csv"
df = pd.read_csv(url)
df.head()
```

As you can notice, the dataset has both categorical and numerical features.

```python
df.dtypes
```

As an example we compute the DCR for the first 1000 rows of the dataset, against the remaining rows. 
In particular, we will compute the 3 nearest neighbors for each row, and we will return both distances and the indexes of the nearest neighbors.

```python
from torch_dcr import dcr

source_df = df.iloc[:1_000]
target_df = df.iloc[len(source_df) :]

dcr_df, indexes_df = dcr(
    source_df=source_df,
    target_df=target_df,
    output_indexes=True,
    k=3,
    metric="l1",
    device="cpu",
    batch_size=100,
)
```
